```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from io import StringIO
```

```python
dados = """Appearance,Temperature,Humidity,Wind,Play
sol,85,85,não,não
sol,80,90,sim,não
nublado,83,86,não,sim
chuva,70,68,não,sim
chuva,68,80,não,
chuva,79,70,sim,sim
menos,64,65,sim,sim
sol,1220,95,não,não
sol,65,70,não,sim
chuva,75,650,não,sim
sol,75,70,sim,sim
nublado,72,90,sim,sim
nublado,81,75,não,sim
chuva,71,91,sim,não
sol,61,92,sim,sim
sol,62,93,sim,sim
sol,63,94,sim,sim
sols,64,95,sim,sim
sols,65,96,sim,sim
sol,66,97,sim,sim"""

df = pd.read_csv(StringIO(dados))

df['Appearance'] = df['Appearance'].replace({'sols': 'sol', 'menos': 'nublado'})

df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')
df['Humidity'] = pd.to_numeric(df['Humidity'], errors='coerce')

df = df[(df['Temperature'] >= 0) & (df['Temperature'] <= 120)]
df = df[(df['Humidity'] >= 0) & (df['Humidity'] <= 100)]

df['Play'] = df['Play'].fillna('não')
df['Wind'] = df['Wind'].fillna('não')

df['Appearance'] = df['Appearance'].astype('category')
df['Wind'] = df['Wind'].astype('category')
df['Play'] = df['Play'].astype('category')
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Appearance')
plt.title('Frequência de Aparência do Tempo')
plt.xlabel('Aparência')
plt.ylabel('Contagem')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='Play', y='Temperature')
plt.title('Temperatura x Jogar')
plt.xlabel('Jogar')
plt.ylabel('Temperatura')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='Play', y='Humidity')
plt.title('Umidade x Jogar')
plt.xlabel('Jogar')
plt.ylabel('Umidade')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Appearance', hue='Play')
plt.title('Aparência vs Jogar')
plt.xlabel('Aparência')
plt.ylabel('Contagem')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Wind', hue='Play')
plt.title('Vento vs Jogar')
plt.xlabel('Vento')
plt.ylabel('Contagem')
plt.show()
```
